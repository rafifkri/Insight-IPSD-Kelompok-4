**tampilkan sales_outlet_id berdasarkan jumlah transaksi tiap outlet**

```python
merged = pd.merge(sr, so, on='sales_outlet_id')
nt = merged.groupby('sales_outlet_id').agg(nilai_transaksi=('transaction_id', 'nunique')).reset_index()
display(nt)
```

**Omset tiap kategory product**

```python
merged = sr.merge(p[["product_id","product_category"]], on="product_id", how="left")
category = merged.groupby("product_category")["line_item_amount"].sum().reset_index()
category
```

**10 product penjualan tertinggi**

```python
top_p = (merged.groupby("product_id")["line_item_amount"].sum().sort_values(ascending=False).head(10).reset_index())
top_p
```

**Omset total**

```python
total = merged["line_item_amount"].sum()
float(total)
```

**Menghitung nilai nunique() pada product dan customer**

```python
n_product = merged["product_id"].nunique()
n_customer = merged["customer_id"].nunique()
n_product, n_customer
```

**KPI harian (tx unik, item, omset)**

```python
kpi = sr.groupby("transaction_date").agg(transaksi_unik=("transaction_id","nunique"),total_item=("quantity","sum"),omzet=("line_item_amount","sum")).reset_index()
kpi.head()
```

**Total penjumlahan tiap produk**

```python
product_sales = sr.groupby('product_id')['line_item_amount'].sum().reset_index()
product_sales = product_sales.sort_values(by='line_item_amount', ascending=False)
product_sales.head(10)
```

**Menampilkan outlet total penjualan tertinggi beserta kota**

```python
outlet_sales_summary = merged.groupby('sales_outlet_id')['line_item_amount'].sum().reset_index()
outlet_sales_summary = pd.merge(outlet_sales_summary, so[["sales_outlet_id","store_city"]], on="sales_outlet_id", how="left")
outlet_sales_summary = outlet_sales_summary.sort_values("line_item_amount", ascending=False)
display(outlet_sales_summary.head(5))
```

**menganalisis stats setiap produk**
* Total pendapatan
* Rata-rata penjualan
* Jumlah unit terjual
* Jumlah transaksi

```python
stats = (merged.groupby(['product_id']).agg(
        total_sales=('line_item_amount', 'sum'),
        avg_sales=('line_item_amount', 'mean'),
        total_qty=('quantity', 'sum'),
        num_transactions=('transaction_id', 'nunique')).reset_index())

stats = stats.sort_values(by='total_sales', ascending=False)
stats.head()
```

**Menganalisa seberapa besar kontribusi tiap produk terhadap total penjualan kategori**

```python
category_summary = merged.groupby("product_category")["line_item_amount"].sum().reset_index().rename(columns={'line_item_amount': 'total_sales', 'product_category': 'category'})
category_summary = category_summary.rename(columns={'total_sales': 'total_sales_category'})
merged_stats = merged.merge(category_summary[['category', 'total_sales_category']], left_on='product_category', right_on='category', how='left')
merged_stats['kontribusi%'] = (merged_stats['line_item_amount'] / merged_stats['total_sales_category']) * 100
merged_stats.head()
```

**outlet dengan penjualan terendah**
